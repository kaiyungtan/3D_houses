## Import libraries

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
```

## Zip file to dataframe

* CSV Flandres  - https://opendata.bosa.be/download/best/openaddress-bevlg.zip
* CSV Région de Bruxelles  - https://opendata.bosa.be/download/best/openaddress-bebru.zip
* CSV Wallonie - https://opendata.bosa.be/download/best/openaddress-bewal.zip

```python
# download and open zip file from a website and then create a dataframe from csv file from the zip file 

def open_zip(url,csv):

    zip_in_memory = urlopen(url).read()

    z = ZipFile(BytesIO(zip_in_memory))
    
    df = pd.read_csv(z.open(csv))
    
    return df
```

```python
# open zip CSV Flandres - https://opendata.bosa.be/download/best/openaddress-bevlg.zip

csv_file_fl = 'openaddress-bevlg.csv'
zip_fl = 'https://opendata.bosa.be/download/best/openaddress-bevlg.zip'
df_flandre = open_zip(zip_fl,csv_file_fl)
```

```python
# open zip CSV Région de Bruxelles - https://opendata.bosa.be/download/best/openaddress-bebru.zip

csv_file_bru = 'openaddress-bebru.csv'
zip_bru = 'https://opendata.bosa.be/download/best/openaddress-bebru.zip'
df_bru = open_zip(zip_bru,csv_file_bru)
```

```python
# open zip CSV Wallonie - https://opendata.bosa.be/download/best/openaddress-bewal.zip

csv_file_wal = 'openaddress-bewal.csv'
zip_wal = 'https://opendata.bosa.be/download/best/openaddress-bewal.zip'
df_wal = open_zip(zip_wal,csv_file_wal)
```

## Flandre Addresses

```python
df_flandre.head()
```

```python
df_flandre.isna().sum()
```

```python
df_flandre['postcode'] = df_flandre['postcode'].astype('str')
```

```python
df_flandre["Address"] = df_flandre["streetname_nl"] + " " + df_flandre["house_number"] + " " +\  
                        df_flandre['postcode'] + " " +  df_flandre['municipality_name_nl']
```

```python
df_flandre_address =  df_flandre[['Address', 'EPSG:31370_x','EPSG:31370_y',
                                  'EPSG:4326_lat','EPSG:4326_lon','region_code']]
```

```python
df_flandre_address.drop_duplicates(subset='Address', keep="first", inplace=True)
```

```python
df_flandre_address.reset_index(drop=True,inplace=True)
```
